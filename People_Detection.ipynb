```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

model = YOLO('/content/runs/detect/train5/weights/best.pt')
image_path = "/content/People-Detection-9/test/images/0000540_jpg.rf.d1b964542ac8df27835f1348ebcac01a.jpg"
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor

image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

results = model(image_rgb)[0]

print("Detected People: ")
for box in results.boxes:
    class_id = int(box.cls[0])
    label = model.names[class_id]
    conf = box.conf[0]
    print(f"{label} {conf:.2f}")

annotated_image = results.plot()
plt.imshow(annotated_image)
plt.axis('off')
plt.title("YOLOv8 Detection")
plt.show()
```
